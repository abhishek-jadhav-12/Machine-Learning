```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet("../data/processed/merged_dataset.parquet")

df.head()
```

```python
feature_df = df.copy()
```

```python
print(feature_df.shape)
```

```python
feature_df["Driver_Experience"] = (
    feature_df.groupby("Driver")["Driver"]
              .transform("count")
)
```

```python
feature_df[["Driver", "Driver_Experience"]].head(10)
```

```python
feature_df["Driver_Experience"].isnull().sum()
```

```python
feature_df[["Driver", "Driver_Experience"]].head(10)
```

```python
feature_df = feature_df.sort_values(
    by=["Season", "Race"]
).reset_index(drop=True)
```

```python
feature_df[["Season", "Race", "Driver"]].head(20)
```

```python
feature_df["Driver_Avg_Finish"] = (
    feature_df.groupby("Driver")["Finish"]
    .transform(lambda x: x.shift().expanding().mean())
)
```

```python
feature_df[
    ["Driver", "Season", "Race", "Finish", "Driver_Avg_Finish"]
].head(20)
```

```python
feature_df["Driver_Avg_Finish"].isnull().sum()
```

```python
feature_df["Driver_Avg_Rating"] = (
    feature_df.groupby("Driver")["Rating"]
              .transform(lambda x: x.shift().expanding().mean())
)
```

```python
feature_df[
    ["Driver", "Rating", "Driver_Avg_Rating"]
].head(20)
```

```python
feature_df["Team_Avg_Rating"] = (
    feature_df.groupby("Team")["Rating"]
              .transform(lambda x: x.shift().expanding().mean())
)
```

```python
feature_df[
    ["Team", "Rating", "Team_Avg_Rating"]
].head(20)
```

```python
feature_df["Make_Avg_Rating"] = (
    feature_df.groupby("Make")["Rating"]
              .transform(lambda x: x.shift().expanding().mean())
)
```

```python
feature_df[
    ["Make", "Rating", "Make_Avg_Rating"]
].head(20)
```

```python
feature_df["Track_Avg_Rating"] = (
    feature_df.groupby("Track")["Rating"]
              .transform(lambda x: x.shift().expanding().mean())
)
```

```python
feature_df[
    ["Track", "Rating", "Track_Avg_Rating"]
].head(20)
```

```python
def classify_track(length):
    if length < 1:
        return "Short Track"
    elif length < 2:
        return "Intermediate"
    else:
        return "Superspeedway"

feature_df["Track_Type"] = feature_df["Length"].apply(classify_track)
```

```python
feature_df[["Length", "Track_Type"]].head(20)
```

```python
def classify_start(pos):
    if pos <= 5:
        return "Front"
    elif pos <= 15:
        return "Mid"
    else:
        return "Back"

feature_df["Start_Category"] = feature_df["Start"].apply(classify_start)
```

```python
feature_df[["Start", "Start_Category"]].head(20)
```

```python
feature_df.columns
```

```python
feature_df.to_parquet(
    "../data/processed/feature_engineered.parquet",
    index=False
)

print("✅ Feature engineered dataset saved successfully!")
```
